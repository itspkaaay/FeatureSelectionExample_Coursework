```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
df= pd.read_csv('/Users/pk/Desktop/Python for DS course/mobile-price-classification/train.csv')
df.head(5)
```

```python
features= df.iloc[:,0:20]
output=df.iloc[:,-1]
features.head(5),output.head(5)
```

```python
bestFeatures=SelectKBest?
```

```python
bestFeatures=SelectKBest(score_func=chi2,k=10)
bestFeatures
```

```python
bestFeatures.fit(features,output)
```

```python
bestFeatures.scores_
```

```python
scores= pd.DataFrame(bestFeatures.scores_)
Variables= df.columns
Variables= pd.DataFrame(Variables)
Relavance_df= pd.concat((Variables,scores),axis=1)
Relavance_df.columns=['Feature','Scores']
Relavance_df=Relavance_df.sort_values('Scores',ascending=False)
Relavance_df
```
